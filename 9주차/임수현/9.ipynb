문제를 보고 dfs,bfs 알고리즘이 아닌 예외처리를 통한 쌩 구현으로 문제를 풀려고 시도함. 실패를 하였고 오랫동안 알고리즘을 활용해서 문제를 안풀다보니 이제는 
알고리즘 구현을 다 까먹게됨. 코테 스터디 문제 풀때만큼은 알고리즘 능력을 열심히 기르자.

```python
# 알고리즘 없이 접근한 풀이
def solution(n, m, x, y, r, c, k):
    # d,l,r,u
    # 가능 불가능 먼저 탐색
    num_lr=y-c
    num_ud=x-r
    if (k-abs(num_lr)-abs(num_ud))%2==1:
        return 'impossible'
    
    # 필수 이동 경로 탐색
    answer=[]
    if num_ud<0: #음수면 아래로 이동해야함
        answer.extend(['d']*abs(num_ud))
    else:
        answer.extend(['u']*abs(num_ud))
    if num_lr<0: #음수면 오른쪽 이동해야함
        answer.extend(['r']*abs(num_lr))
    else:
        answer.extend(['l']*abs(num_lr))
    
    answer.sort()
    # 딱 필수 이동만큼만 이동해야하는 경우
    if (k-abs(num_lr)-abs(num_ud))==0:
        return ''.join(answer)
    # normal한 경우에 변곡점에서 이벤트!
    def check_down_moving(loc):
        a=loc[0]
        b=loc[1]
        if n-a>0:
            return n-a
        else:
            return 0
    def check_left_moving(loc):
        a=loc[0]
        b=loc[1]
        return b
    
    # 변곡점 탐색
    prev=0
    datas=[]
    count=0
    for move in answer:
        if prev!=move:
            if count>0:
                datas.append([prev,count])
            prev=move
            count=1
        else:
            count+=1
    datas.append([prev,count])
    print(datas)
    # 재구성
    now=[x,y]
    coin = int((k-abs(num_lr)-abs(num_ud))/2)
    result=[]
    for data in datas:
        if data[0]!='d' and check_down_moving(now):
            iter=check_down_moving(now)
            if iter<coin:
                coin-=iter
                result.extend()
    return answer
```

```python
# 블로그를 참고해서 코드 동일하게 했는데 올솔이 아님. 뭐지?
dx = [1, 0, 0, -1]
dy = [0, -1, 1, 0]
dAlpha = ['d', 'l', 'r', 'u']
answer = "z"

def isValid(nx,ny,n,m):
    return 1<=nx<=n and 1<=ny<=m

def dfs(n,m,x,y,r,c,prev_s,cnt,k):
    global answer
    # 탐색 과정에서 이동할 수 있는 거리 부족시 스탑
    if k<cnt+abs(x-r)+abs(y-c):
        return
    # 목표 지점 도달시 스탑
    if x == r and y == c and cnt == k:
        answer = prev_s
        return 
    for i in range(4):
        if isValid(x + dx[i], y + dy[i], n, m) and prev_s < answer:
            dfs(n,m,x+dx[i],y+dy[i],r,c,prev_s+dAlpha[i],cnt+1,k)
    # prev_s < answer로 문자열 사전적 우선순위를 판별할 수 있다.
            
def solution(n,m,x,y,r,c,k):
    dist=abs(x-r)+abs(y-c)
    if dist>k or (k-dist)%2==1:
        return "impossible"
    dfs(n,m,x,y,r,c,"",0,k)
    return answer
```
